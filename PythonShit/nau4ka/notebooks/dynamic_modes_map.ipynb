```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
import pandas as pd
import seaborn as sb


def print_from_csv(name, maxpar, cmap="tab20", k=10):
    path = "C:\\Users\\Lenovo\\PycharmProjects\\nau4ka\\" + name
    df = pd.read_csv(path)

    h = df.shape[0]
    h_ = int(h/(k-1))
    space_par = np.linspace(0, maxpar, h)

    val = list(range(0, h, h_))
    lab1 = [round(space_par[i], 2) for i in range(0, h, h_)]
    lab2 = [round(space_par[h - 1 - i], 2) for i in range(0, h, h_)]

    sb.heatmap(df, cmap=cmap)
    
    plt.xticks(ticks=val, labels=lab1)
    plt.yticks(ticks=val, labels=lab2)
    plt.show()
```

```python
plt.xlabel("Параметр $а$")
plt.ylabel("Параметр $\\rho$")
plt.title("$v=0.5$, $s=0.5$, $(x_0, y_0)=(3,3)$")

print_from_csv("dmm1.csv", 80)
```
