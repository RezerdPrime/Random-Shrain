```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

#plt.style.use('dark_background')

def f(v, s, a, rho, xy):
    return [a * xy[1], (s*xy[0] + v*xy[1])*exp(-xy[0] - rho*xy[1])]


# Биф диаграмма по параметру rho
def BifurcationDiagram1(v, s, a, Rs, xy, n=200, printable_coord=False, size=1, width=1):
    l = len(Rs)
    bufcurve = [f(v, s, a, r, xy) for r in Rs]
    
    for _ in range(1500):
        bufcurve = [f(v, s, a, Rs[i], bufcurve[i]) for i in range(l)]

    for _ in range(n):
        bufcurve = [f(v, s, a, Rs[i], bufcurve[i]) for i in range(l)]
        #plt.plot(Rs, [p[printable_coord] for p in bufcurve], linestyle="dotted", color="blue", linewidth=linewidth)
        plt.scatter(Rs, [p[printable_coord] for p in bufcurve], c=[[0, 0, 1, 0.1]], s=size, marker=".", linewidths=width)


# Биф диаграмма по параметру a
def BifurcationDiagram2(v, s, As, r, xy, n=200, printable_coord=False, size=1, width=1):
    l = len(As)
    bufcurve = [f(v, s, a, r, xy) for a in As]
    
    for _ in range(1500):
        bufcurve = [f(v, s, As[i], r, bufcurve[i]) for i in range(l)]

    for _ in range(n):
        bufcurve = [f(v, s, As[i], r, bufcurve[i]) for i in range(l)]
        #plt.plot(As, [p[printable_coord] for p in bufcurve], linestyle="dotted", color="blue")
        plt.scatter(As, [p[printable_coord] for p in bufcurve], c=[[0, 0, 1, 0.1]], s=size, marker=".", linewidths=width)
```

```python
v = 0.5
s = 0.5
n = 200
r = 64
d = 2000

xy = [3,3]

As = np.linspace(0, 80, d)

plt.figure(figsize=(8, 5))
BifurcationDiagram2(v, s, As, r, xy, size=0.5, width=1)
plt.show()
```
