```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

def f(v, s, a, rho, xy):
    return [a * xy[1], (s*xy[0] + v*xy[1])*exp(-xy[0] - rho*xy[1])]


def BifurcationDiagram(v, s, xy, fa, fr, begin=0, end=1, d=10, n=200, printable_coord=False, size=1, width=1):

    ts = np.linspace(begin, end, d)
    Rs = [fr(t) for t in ts]
    As = [fa(t) for t in ts]

    bufcurve = [f(v, s, As[i], Rs[i], xy) for i in range(d)]
    
    for _ in range(1500):
        bufcurve = [f(v, s, As[i], Rs[i], bufcurve[i]) for i in range(d)]

    for _ in range(n):
        bufcurve = [f(v, s, As[i], Rs[i], bufcurve[i]) for i in range(d)]
        plt.scatter(ts, [p[printable_coord] for p in bufcurve], c=[[0, 0, 1, 0.1]], s=size, marker=".", linewidths=width)
```

```python
v = 0.5
s = 0.5
n = 200
d = 2000

xy = [3,3]

As = np.linspace(0, 80, d)

def fa(t): return t * 0.589788025031
def fr(t): return t * 0.807558100405 + 5.1

plt.figure(figsize=(5, 4))
#(v, s, xy, fa, fr, begin=0, end=80, d=d, n=n, size=0.5, width=0.75)
BifurcationDiagram(v, s, xy, fa, fr, begin=0, end=80, d=d, n=n, size=0.25, width=1)
plt.show()
```
